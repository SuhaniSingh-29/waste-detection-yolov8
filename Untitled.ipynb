### Real-time detection using YOLOv8

```python
import cv2
from ultralytics import YOLO
import random
```

#### Loading the Trained YOLOv8 Model

```python
model = YOLO("runs/detect/train23/weights/best.pt")
```

```python
# Defining class-wise colors
colors = {
    'battery': (255, 0, 0),
    'metal': (192, 192, 192),
    'glass': (0, 255, 255),
    'shoes': (139, 69, 19),
    'biological': (0, 255, 0),
    'paper': (255, 255, 0),
    'plastic': (255, 0, 255),
    'cardboard': (210, 105, 30),
    'clothes': (138, 43, 226),
    'trash': (128, 128, 128),
    'object': (100, 149, 237),
    'background': (105, 105, 105)
}
```

```python
# Open webcam (use 0 for built-in or 1 for external camera)
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()
```

```python
while True:
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame.")
        break

    results = model(frame, conf=0.3)

    for result in results:
        boxes = result.boxes
        for box in boxes:
            x1, y1, x2, y2 = box.xyxy[0]
            conf = float(box.conf[0])
            cls_id = int(box.cls[0])
            class_name = model.names[cls_id]

            color = colors.get(class_name, (0, 0, 255))

            # Bounding box
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), color, 2)
            label = f"{class_name} {conf:.2f}"
            cv2.putText(frame, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

    cv2.imshow('YOLOv8 Waste Detection (Real-Time)', frame)

    #Press button 'q' to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
```

```python
# Release resources
cap.release()
cv2.destroyAllWindows()
```
